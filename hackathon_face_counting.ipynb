we will use dlib cnn model for detecting faces which is state of art model

```python
x1=pd.read_csv('/content/drive/MyDrive/project_face_detector/train.csv')#reading throught the train data
```

```python
#creating empty list to store predictions
x2=[]
```

```python
import cv2                                                                        #importing cv2 , dlib,imutils 
import dlib
from imutils import face_utils
for i in tqdm(range(x1.shape[0])):                                                  #starting the loop with range shape of x1
    z2 = '/content/drive/MyDrive/project_face_detector/image_data/'+ x1['Name'][i]   #creating path for each train image 
    #print(z1)
    #break   
    frame1 = cv2.imread(z2)                                                      #reading the image through cv2 imread function
    frame1 = cv2.cvtColor(frame1, cv2.COLOR_BGR2RGB)
    gray1 = cv2.cvtColor(frame1, cv2.COLOR_BGR2GRAY)                             #preprocessing the image   

    dnnFaceDetector1 = dlib.cnn_face_detection_model_v1("/content/drive/MyDrive/project_face_detector/mmod_human_face_detector.dat")
    rects1 = dnnFaceDetector1(gray1, 1)                                          #appling the pretrained state of art model which is trained earlier on huge dataset    
    x2.append(len(rects1))                                                       #predicting the image using this model and appending it to list 
```

```python
data1=pd.read_csv('/content/drive/MyDrive/project_face_detector/train.csv')
```

```python
#now making a dataframe of predictions and adding it to train data
data1['dlib_cnn_train_results']=pd.DataFrame(x2)
```

```python
#checking the accuracy score of model
from sklearn.metrics import accuracy_score
Training_accuracy_dlib=accuracy_score(data1['HeadCount'],data1['dlib_cnn_train_results'])
```
